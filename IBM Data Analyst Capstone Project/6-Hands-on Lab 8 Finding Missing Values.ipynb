### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
print(df.info())
df.dtypes
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
#True = missing data
print(df.isnull())


missing_cols = df.columns[df.isnull().any()]
print(missing_cols)
```

```python
df.isnull().sum()
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure (figsize=(10,6))


sns.heatmap(
    df.isnull(),
    cmap="coolwarm",
    yticklabels=False,   # hide row labels (optional)
    cbar=True            # show color bar
)

plt.title("Missing Data Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
print(df.isnull().sum())
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
Majority_Value=df['Employment'].mode()[0]

print('Most Frequent value:', Majority_Value)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df['Employment'] = df['Employment'].fillna("full-time")
print(df['Employment'].isnull().sum())
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
#Bar graph for visulization, but unable to read and not clear,
counts = df["Employment"].value_counts()

ax = counts.plot(kind="bar", figsize=(10, 5))
ax.set_title("Employment distribution (after imputation)")
ax.set_xlabel("Employment")
ax.set_ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#Changed to a Horizontal Bar Chart which allows you to read the data more clearly.

counts = df["Employment"].value_counts()

# Keep only top N categories, group the rest into "Other"
N = 10
top = counts.head(N)
other = counts.iloc[N:].sum()
if other > 0:
    top.loc["Other"] = other

# Plot horizontal bars (readable labels)
ax = top.sort_values().plot(kind="barh", figsize=(10, 6), color="Green") 
ax.set_title(f"Employment distribution after imputation (Top {N} + Other)")
ax.set_xlabel("Count")
ax.set_ylabel("Employment")
plt.tight_layout()  
plt.show()
```
