#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
## Write your code here
df = df[['YearsCodePro', 'JobSat']].dropna()

#convert YearsCodePro to number -  
def convert_years(x):
    x=str(x)
    if 'Less than' in x:
        return 0.5
    elif 'More than' in x:
        return float(x.split()[2]) +1
    else:
        return pd.to_numeric(x, errors='coerce')
        
df['YearsCodePro_num'] = df['YearsCodePro'].apply(convert_years)

df["JobSat_num"] = pd.to_numeric(df["JobSat"], errors="coerce")

df = df.dropna(subset=["YearsCodePro_num", "JobSat_num"])


bins = [0, 5, 10, 20, float('inf')]
labels = ['0-5', '5-10', '10-20', '>20']

df['experience_range'] = pd.cut(df['YearsCodePro_num'], bins=bins, labels=labels, right=False)

median = df.groupby('experience_range')['JobSat_num'].median()

# Plot Bar
ax=median.plot(kind='bar', title = 'Median JobSat by YearsCodePro Range',)
ax.set_xlabel('YearsCodePro range (Years)')
ax.set_ylabel('Median JobSat')
plt.tight_layout()
plt.show()
```

```python
#Need to clean Employment data - map unreadble - 
for col in ['RemoteWork', 'Employment', 'MainBranch']:
 df[col] = (df[col].astype(str)
               .str.strip()
               .str.replace(r'\s+', ' ', regex=True))


  # shorten MainBranch Labels, make more precise - 
mainbranch_map = {
    'I am a developer by profession': 'Dev (Professional)',
    'I used to be a developer by profession, but no longer am': 'Ex-Dev (Previously)',
    'I am not primarily a developer, but I write code sometimes as part of my work/studies' :'Pt-Dev, (occasionally)',
    'I code primarily as a hobby': 'Hobby-Dev',
    'I am learning to code': 'Learner-Dev'}

df['MainBranch_tidy'] = df['MainBranch'].map(mainbranch_map).fillna(df['MainBranch'])

#Shorten Employment Labels and tidy - 
df['Employment_clean'] =( df['Employment']
.str.replace('Employed, full-time', 'Full-time Employed', regex=False)
.str.replace('Employed,part-time', 'Part-time Employed', regex=False)
.str.replace('Student,full-time', 'Full-time Student', regex=False)
.str.replace('Student,part-time', 'Part-time Student', regex=False)
.str.replace('Independentcontractor,freelancer,orself-employed', 'Self-employed / Contractor', regex=False)
.str.replace('Notemployed,butlookingforwork', 'Seeking Employment', regex=False)
.str.replace('Notemployed,andnotlookingforwork', 'Not-Seeking Employment', regex=False)
.str.replace('Retired', 'Retired', regex=False))

def simp_employment(emp):
    emp = str(emp)
    
    if 'Full-time Employed' in emp:
        return 'Full-time Employed'
    elif 'Part-time Employed' in emp:
        return 'Part-time Employed'
    elif 'Self-employed / Contractor' in emp:
        return 'Self-employed / contractor'
    elif 'Full-time Student' in emp or 'Part-time Student' in emp:
        return 'Student'
    elif 'Seeking Employment' in emp:
        return 'Seeking Employment'
    elif 'Not-Seeking Employment' in emp:
        return 'Not-Seeking Employment'
    elif 'Retired' in emp:
        return 'Retired'
    else:
        return 'Other / Mixed'

df['Employment_tidy'] = df['Employment_clean'].apply(simp_employment)

 
#Percentages calc.
crosstab_pct = pd.crosstab(
    index = [df['Employment_tidy'], 
df['MainBranch_tidy']],
    columns = df['RemoteWork'],
    normalize='index')*100

#HeatMap plot
plt.figure(figsize=(15,10))
sns.heatmap(crosstab_pct, annot = True,
            fmt='.1f', cmap='YlGnBu')
plt.title('Remote Work Preferences by Employment Type and Developer Backgroung (%)')
plt.xlabel('RemoteWork')
plt.ylabel('Employment Type and Developer')
plt.tight_layout()
plt.show()
```

```python
#Cleaning - 
df_lang = df[['Country', 'LanguageHaveWorkedWith']].dropna()

#Standardise
df_lang['Country'] = df_lang['Country'].astype(str).str.strip()
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].astype(str).str.strip()

#Filtering data by the country - 
selected_country = 'United States of America'
df_country = df_lang[df_lang['Country'] == selected_country].copy()

df_country = df_lang[df_lang['Country'] == selected_country].copy()
print(f'Rows for {selected_country}:{len(df_country)}')

#splitting into seperate rows
lang_exploded = (df_country.assign(
    LanguageHaveWorkedWith=df_country['LanguageHaveWorkedWith'].str.split(';')).explode('LanguageHaveWorkedWith'))

#clean more after exploding
lang_exploded['LanguageHaveWorkedWith'] = lang_exploded['LanguageHaveWorkedWith'].str.strip()
lang_exploded=lang_exploded.dropna(subset=['LanguageHaveWorkedWith'])

#Top 10 Languages
top_n = 10
lang_counts = lang_exploded['LanguageHaveWorkedWith'].value_counts().head(top_n)

print(lang_counts)
```

```python
#Create a subset
plot_df = df[['YearsCodePro', 'JobSatPoints_1']].copy()

#clean yearscodepro
plot_df['YearsCodePro'] = plot_df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51})

#ensure numerical values
plot_df['YearsCodePro'] = pd.to_numeric(plot_df['YearsCodePro'], errors='coerce')
plot_df['JobSatPoints_1'] = pd.to_numeric(plot_df['JobSatPoints_1'], errors='coerce')

#Remove NAN 
plot_df = plot_df.dropna()

#correlation
corr = plot_df['YearsCodePro'].corr(plot_df['JobSatPoints_1'])
print(f'Pearson Correlation: {corr:.3f}')

# Create a scatter plot
plt.figure(figsize=(10,6))

sns.scatterplot(
    data = plot_df,
    x = 'YearsCodePro',
    y = 'JobSatPoints_1',
    alpha = 0.3,
    s=20)

#adding a trend line
sns.regplot(
    data = plot_df,
    x = 'YearsCodePro',
    y = 'JobSatPoints_1',
    scatter = False,
    color = 'green')

plt.title('Years of Experience vs  Job Satisfaction')
plt.xlabel('YearsCodePro')
plt.ylabel('JobSatPoints_1')
plt.tight_layout()
print(plot_df.shape)
plt.show()
```
