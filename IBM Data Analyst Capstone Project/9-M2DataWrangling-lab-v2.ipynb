### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

```python
# Write your code here
#Column data types - 
print(df.dtypes)

missing_values=df.isnull()
print(missing_values)

missing_count = df.isnull().sum()
print("Number of missing values:", missing_count)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
irrelevent_entries = df.isnull()
print(irrelevent_entries.head())

for Country in irrelevent_entries.columns.values.tolist():
    print(Country)
    print(irrelevent_entries[Country].value_counts())
    print("")
    
```

```python
Specific_Columns = ['Country']

for col in Specific_Columns:
    print(col)
    print(irrelevent_entries[col].value_counts())
    print("")
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
invalid_values = ['n/a', 'unknown', '', 'none' 'NaN', 'NA', 'nan']

mask = df['Country'].str.lower().isin(invalid_values)
print(df[mask])

print(df['Country'].value_counts(dropna=False))
```

```python
df['Country_clean'] = df['Country'].astype('string').str.strip().str.lower()
print(df['Country_clean'])

country_list=df['Country_clean'].unique()
print(country_list)

Country_mapping = Country_mapping = {country: country for country in country_list}
print(Country_mapping)
#Mapping
df['Country_mapping'] = df['Country_clean'].map(Country_mapping)

df['Country_mapping'] = df['Country_mapping'].fillna(df['Country_clean'])

print(df['Country_mapping'].unique())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df['Employment'].unique()

one_hot = pd.get_dummies(df['Employment'],prefix='Employment')

df_encoding = pd.concat([df, one_hot], axis=1)

print(df_encoding)
        
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values = df.isnull().sum()

missing_values_desc = missing_values.sort_values(ascending=False)

print(missing_values_desc.head())
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here

df = df.fillna(df.mean(numeric_only=True))
print(df.isna().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
most_frequent = df['RemoteWork'].mode()[0]
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent)
print(df.isna().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
!pip install scikit-learn
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_normalized=scaler.fit_transform(df[['ConvertedCompYearly']])
df['ConvertedCompYearly'] = data_normalized.flatten()


print(df['ConvertedCompYearly'].min())  # should be 0
print(df['ConvertedCompYearly'].max())  # should be 1

print(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
print(df['YearsCodePro'].max())
```

```python
## Write your code here
#Binning

df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51})

df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

bins = [0, 4, 8, float('inf')]
labels = ['Beginner', 'Intermediate', 'Pro']

df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=bins,
labels=labels, include_lowest=True)

print(df['ExperienceLevel'].value_counts())
print(df.head())
```
